**3. Customer Lifetime Value (CLV)**

**4. Market Reputation and Trust**
Frequent customer exits can reflect poorly on a bank's service quality. Lower churn rates indicate higher customer satisfaction and trust.

**5. Predictive Strategy and Personalization**
Understanding churn behavior enables banks to proactively intervene with at-risk customers through personalized offerings, loyalty rewards, or improved service quality.

##üîç Understanding the F1 Score, Macro Avg, and Weighted Avg##

**‚úÖ F1 Score:**

-F1 Score is the harmonic mean of precision and recall. It's especially useful in imbalanced datasets (like churn prediction).

-It gives a single score to balance false positives and false negatives.

**üìä Macro Average:**

-This takes the simple average of the F1 scores of both classes (churn and non-churn), without considering class imbalance.

-Macro F1 Score = 0.72 ‚Üí Indicates moderate overall performance across both classes equally.

**‚öñÔ∏è Weighted Average:**

-This considers the support (number of instances) for each class and calculates a weighted average of F1 scores.

-Weighted F1 Score = 0.84 ‚Üí Since class 0 (non-churners) dominates the data, the overall score is high due to better performance on the majority class.

**üìà Managerial Implications**

**1. Imbalanced Focus**

-The model performs very well for non-churners (F1 = 0.91), but less accurately for churners (F1 = 0.53, recall = 0.41).

-‚ùó This means 59% of churners are being missed, which is critical for a business aiming to reduce churn.

‚Üí Action: Invest in techniques like SMOTE, cost-sensitive learning, or resampling to improve recall for churners.

##üîç  ROC-AUC Score##
ROC-AUC stands for **Receiver Operating Characteristic** - Area Under the Curve.

ROC Curve: Plots the True Positive Rate (Recall) vs. False Positive Rate at various threshold levels.

AUC Score (Area Under Curve): Measures the model‚Äôs ability to distinguish between the two classes.

AUC = 1.0 ‚Üí Perfect classifier
AUC = 0.5 ‚Üí Random guessing
AUC = 0.8527 ‚Üí Excellent model with ~85% probability that it ranks a random churner higher than a non-churner.

##Preprocess the Data##

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd

# Copy the original dataframe
data = df.copy()

# Drop customer_id
data.drop('customer_id', axis=1, inplace=True)

# Label encode gender
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])

# One-hot encode 'country'
data = pd.get_dummies(data, columns=['country'], drop_first=True)

# Features and target
X = data.drop('churn', axis=1)
y = data['churn']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=59)
```

##Build and Train the ANN Model##

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Build the ANN
model = Sequential()

# Input layer + First hidden layer
model.add(Dense(units=16, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))

# Second hidden layer
model.add(Dense(units=12, activation='relu'))
model.add(Dropout(0.3))

# Output layer
model.add(Dense(units=1, activation='sigmoid'))  # Binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32)
```

##Visualize Accuracy and Loss##

```python
import matplotlib.pyplot as plt

# Plot Accuracy
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='green')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', color='red')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

## Evaluation & Interpretation##

**Accuracy on test data**

```python
# Evaluate accuracy on test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

**Confusion Matrix & Classification Report**

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Predict classes (0 or 1)
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**ROC-AUC Score**

```python
from sklearn.metrics import roc_auc_score

roc_score = roc_auc_score(y_test, y_pred_prob)
print(f"ROC-AUC Score: {roc_score:.4f}")
```

```python
import keras_tuner as kt
from tensorflow.keras import layers

def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Dense(units=hp.Int('units_input', min_value=16, max_value=128, step=16), activation='relu', input_dim=X_train.shape[1]))
    model.add(layers.Dense(units=hp.Int('units_hidden', min_value=16, max_value=128, step=16), activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(
            hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=2,
    directory='churn_tuning',
    project_name='bank_churn_ann'
)

tuner.search(X_train, y_train, epochs=20, validation_split=0.2, verbose=1)
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
Best Hyperparameters:
- Input Layer Units: {best_hps.get('units_input')}
- Hidden Layer Units: {best_hps.get('units_hidden')}
- Learning Rate: {best_hps.get('learning_rate')}
""")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras_tuner as kt
```

```python
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Dense(units=hp.Int('units_input', min_value=16, max_value=128, step=16),
                           activation='relu', input_dim=X_train.shape[1]))
    model.add(layers.Dense(units=hp.Int('units_hidden', min_value=16, max_value=128, step=16),
                           activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(
            hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=2,
    directory='churn_tuning',
    project_name='bank_churn_ann'
)

tuner.search(X_train, y_train, epochs=20, validation_split=0.2, verbose=1)
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"""
üîß Best Hyperparameters:
- Input Layer Units: {best_hps.get('units_input')}
- Hidden Layer Units: {best_hps.get('units_hidden')}
- Learning Rate: {best_hps.get('learning_rate')}
""")
```

```python
# Build model with best hyperparameters
model = tuner.hypermodel.build(best_hps)

# Train with best setup
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32)
```
